# Merging these three shades

```python
img = []
for i in range(256):
    t = []
    for j in range(256):
        t.append([0, 0, i])
    img.append(t)

for i in range(256, 0, -1):
    t = []
    for j in range(256):
        t.append([0, 0, i])
    img.append(t)

img = np.array(img)
```

```python
plt.imshow(img)
```

```python
img = []
for i in range(256):
    t = []
    for j in range(256):
        t.append([i, 0, 0])
    img.append(t)


for i in range(256, 0, -1):
    t = []
    for j in range(256):
        t.append([0, 0, i])

    img.append(t)
img = np.array(img)
```

```python
plt.imshow(img)
```

```python
img = []
for i in range(256):
    t = []
    for j in range(1536):
        t.append([i, 0, 0])
    img.append(t)
for i in range(256, 0, -1):
    t = []
    for j in range(1536):
        t.append([i, 0, 0])
    img.append(t)
for i in range(256):
    t = []
    for j in range(1536):
        t.append([0, i, 0])
    img.append(t)
for i in range(256, 0, -1):
    t = []
    for j in range(1536):
        t.append([0, i, 0])
    img.append(t)
for i in range(256):
    t = []
    for j in range(1536):
        t.append([0, 0, i])
    img.append(t)
for i in range(256, 0, -1):
    t = []
    for j in range(1536):
        t.append([0, 0, i])
    img.append(t)

img = np.array(img)

# Reshape the img array to have the correct shape
# img = img.reshape((256 + 256 + 256 + 256 + 256 + 256, 180 + 1080 + 1080 + 1080 + 1080 + 1080, 3))


```
