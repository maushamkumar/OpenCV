```python
# Create a figure and axis
fig, ax = plt.subplots()

# Create a circle
circle = plt.Circle((0.5, 0.5), 0.3, color='blue', fill=False)

# Add the circle to the axis
ax.add_artist(circle)

# Set the aspect of the plot to be equal
ax.set_aspect('equal', adjustable='datalim')

# Display the plot
plt.show()
```

```python
# Image size
size = 100

# Create a grid of coordinates
x, y = np.meshgrid(np.linspace(-1, 1, size), np.linspace(-1, 1, size))

# Calculate the distance from the center
radius = np.sqrt(x**2 + y**2)

# Create a circle mask (1 inside the circle, 0 outside)
circle_mask = np.where(radius <= 1, 1, 0)

# Create a circle with different color values
circle = np.zeros((size, size, 3))
circle[:, :, 0] = circle_mask  # Red channel
circle[:, :, 1] = circle_mask  # Green channel
circle[:, :, 2] = circle_mask  # Blue channel

# Display the circle
plt.imshow(circle)
plt.axis('off')  # Turn off axis
plt.show()
```
